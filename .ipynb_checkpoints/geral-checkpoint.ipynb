```python
import pandas
import numpy as np
from pandas.tools import plotting

data = pandas.read_csv("data/RLS5_csv.csv", sep=",")
                                                                           
data
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(data.Tmax, data.Tmin)
plt.show()
```

```python
#Kosmogorov Smirnov P
from scipy import stats
p = data.P.copy()

m = np.mean(p)
std = np.std(p)

print m
print std

p = np.array([(i - m)/std for i in sorted(list(p))] )

plt.plot(p)
plt.show()
print p

print stats.kstest(p, 'norm', N=p.size)

"G(x)>=F(x), portanto nao e normal"
```

```python
#Kosmogorov Smirnov Tmax
from scipy import stats
Tmax = data.Tmax.copy()

m = np.mean(Tmax)
std = np.std(Tmax)

print m
print std

Tmax = np.array([(i - m)/std for i in sorted(list(Tmax))] )

plt.plot(Tmax)
plt.show()
print Tmax

print stats.kstest(Tmax, 'norm', mode='asymp', N=Tmax.size)
print "G(x)>=F(x), portanto nao e normal"
```

```python
#Kosmogorov Smirnov Tmax
from scipy import stats
Tmin = data.Tmax.copy()

m = np.mean(Tmin)
std = np.std(Tmin)

print m
print std

Tmin = np.array([(i - m)/std for i in sorted(list(Tmin))] )

plt.plot(Tmin)
plt.show()
print Tmax

print  stats.kstest(Tmin, 'norm', mode='asymp', N=Tmax.size)
print "G(x)>=F(x), portanto nao e normal"
```

```python
from sklearn import linear_model

y_train = np.array(data.Tmax)[0:-200]
y_test = np.array(data.Tmax)[-200:]

x_train = np.array(data.Tmin)[0:-200].reshape(-1, 1)
x_test = np.array(data.Tmin)[-200:].reshape(-1, 1)

regr = linear_model.LinearRegression()

regr.fit(x_train, y_train)

#coeficientes
print('Coefficients: \n', regr.coef_)

#MSE
print("Mean squared error: {}"
      % np.mean((regr.predict(x_test) - y_test) ** 2))

#R2
print("R2: {}".format(regr.score(x_test, y_test)))

# Plot outputs
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, regr.predict(x_test), color='blue',
         linewidth=3)

# plt.xticks(())
# plt.yticks(())

plt.show()
```

```python
#anova
from sklearn.feature_selection import f_classif
print f_classif(x_test, y_test)

```

```python
#outra maneira
import statsmodels.formula.api as sm
# regression_2
regr_2 = sm.OLS(x_train, y_train).fit()
regr_2.summary()
```

```python
import scipy

N = regr_2.nobs
P = regr_2.df_model
dfn, dfd = P, N - P - 1
F = regr_2.mse_model / regr_2.mse_resid
p = 1.0 - scipy.stats.f.cdf(F,dfn,dfd)
print 'F-statistic: {:.3f},  p-value: {:.5f}'.format( F, p )
```
